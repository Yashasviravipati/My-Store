**1, Importing the dataset file**

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

a, Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

```python
print("Missing values before handling:")
print(df.isnull().sum())
```

In this dataset, missing values in the Mileage, Engine, and Power columns were imputed using the median since they are continuous numerical features and median is more robust to outliers than mean. The Seats column, being a categorical numerical feature, was imputed using the mode, as it represents the most frequently occurring value and maintains consistency. The New_Price column was dropped due to over 83% missing values, making imputation unreliable and potentially misleading. These actions ensure data completeness while preserving accuracy and reducing bias in the analysis.

```python
df.drop('New_Price', axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
print("Missing values after cleaning:")
print(df.isnull().sum())
```

b, Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
# Convert the relevant columns to string type before using .str accessor
df['Mileage'] = df['Mileage'].astype(str).str.replace(r'\s+[a-zA-Z/]+', '', regex=True)
df['Engine'] = df['Engine'].astype(str).str.replace(r'\s+CC', '', regex=True)
df['Power'] = df['Power'].astype(str).str.replace(r'\s+bhp', '', regex=True)
df.head()
# Convert 'Mileage', 'Engine', and 'Power' columns to numeric
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')
```

Replace the missing values.

```python
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)
df['Engine'].fillna(df['Engine'].median(), inplace=True)
df['Power'].fillna(df['Power'].median(), inplace=True)
df['Seats'].fillna(df['Seats'].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.head()
```

c, Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python
fuel_mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
df['Fuel_Type'] = df['Fuel_Type'].replace(fuel_mapping)
```

```python
transmission_mapping = {'Manual': 0, 'Automatic': 1}
df['Transmission'] = df['Transmission'].replace(transmission_mapping)
```

d, Create one more feature and add this column to the dataset (you can use mutate function in
R for this). For example, you can calculate the current age of the car by subtracting “Year” value
from the current year.

```python
from datetime import datetime
current_year = datetime.now().year
df['Car_Age'] = current_year - df['Year']
```
